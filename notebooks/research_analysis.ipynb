```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sys
import os

# Add root folder to path to import local modules
sys.path.append(os.path.abspath('..'))

from src.data_loader import DataLoader
from src.config import DAYS

# 1. Load Data
loader = DataLoader()
df = loader.generate_ran_traffic(days=DAYS)

print("Data successfully generated.")
display(df.head())
display(df.describe())

# 2. Distribution Analysis
plt.figure(figsize=(12, 5))
plt.hist(df['throughput_mbps'], bins=50, color='skyblue', edgecolor='black')
plt.title('Throughput Distribution (Histogram)')
plt.xlabel('Mbps')
plt.ylabel('Frequency')
plt.grid(True, alpha=0.3)
plt.show()

# 3. Time Series Analysis (First 7 days)
plt.figure(figsize=(15, 6))
plt.plot(df['throughput_mbps'][:24*7], label='Throughput (Mbps)', color='orange')
plt.title('Weekly Traffic Pattern Analysis')
plt.xlabel('Hours')
plt.ylabel('Mbps')
plt.legend()
plt.grid(True)
plt.show()

print("ANALYSIS CONCLUSION:")
print("1. Strong daily seasonality observed (Day/Night cycle).")
print("2. Anomalies detected (spikes) which the LSTM must learn to handle.")
```
